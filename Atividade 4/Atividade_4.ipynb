# A - Correlação

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# pontos rosas y varia de 1 a 9 x fixo em 0
x_pink = np.zeros(9)  
y_pink = np.linspace(1, 9, 9)

# pontos verdes x varia de 1 a 9 y fixo em 0
x_green = np.linspace(1, 9, 9) 
y_green = np.zeros(9)  


# Regressão linear aplicada aos pontos azuis
slope, intercept, r_value, p_value, std_err = linregress(x_green, y_pink)
# slope * x_green + intercept, está gerando um array de valores y para cada ponto de x_green, que corresponde 
# a pontos ao longo da reta de regressão.
regression_line = slope * x_green + intercept  


fig, ax = plt.subplots(figsize=(8, 6))
# Plotando os pontos rosa
ax.scatter(x_pink, y_pink, color='pink', edgecolor='k', s=200, label='Gene Y variando (x=0)')
# Plotando os pontos verdes
ax.scatter(x_green, y_green, color='lightgreen', edgecolor='k', s=200, label='Gene X variando (y=0)')


# Plotando os pontos azuis: relação linear entre Gene X e Gene Y
ax.scatter(x_green, y_pink, color='skyblue', edgecolor='k', s=200, label='Cruzamento (x,y)')
# Plotando a reta de regressão baseada nos pontos azuis
ax.plot(x_green, regression_line, 'k--', label='Regressão linear')

ax.set_xlabel('Gene X', fontsize=14, weight='bold')
ax.set_ylabel('Gene Y', fontsize=14, weight='bold')
ax.legend()
plt.tight_layout()  # Ajusta layout para não cortar texto
plt.show()  # Exibe o gráfico

# covariancia
std_x = np.std(x_green, ddof=1)  # ddof=1 para amostra
std_y = np.std(y_pink, ddof=1)
cov = r_value * std_x * std_y

# Exibindo resultados da regressão
print(f"Inclinação: {slope:.2f}")
print(f"Intercepto: {intercept:.2f}")
print(f"Coeficiente de correlação (r): {r_value:.2f}")
print(f"P-valor: {p_value:.2e}")
print(f"Erro padrão: {std_err:.2f}")
print("Covariância:", cov)
```

A inclinação de 1.00 significa que, sempre que o valor de x aumenta 1 unidade, o valor de y também aumenta exatamente 1 unidade.

O intercepto de 0.00 mostra onde a reta corta o eixo y quando x=0.

O coeficiente de correlação (r) igual a 1.00 indica uma relação perfeitamente linear e positiva.

O p-valor é extremamente pequeno (3.29 × 10⁻⁷⁰), o que confirma que essa relação não aconteceu por acaso (é altamente significativa).

O erro padrão é 0.00 o que significa que não há incerteza na inclinação da reta 

Covariância é 10.14 indica uma relação positiva

# B: Estimando um Intervalo de Confiança para a Média

```python
from scipy.stats import t
import numpy as np

# Gerar dados amostrais entre 0 e 1 (notas de qualidade visual normalizadas)
np.random.seed(42)
amostra = np.random.beta(a=5, b=2, size=30)

def calcular_intervalo_confianca(amostra, nivel_confianca):
    # Calcular média e desvio padrão amostral
    media = np.mean(amostra)
    desvio_padrao = np.std(amostra, ddof=1)
    n = len(amostra)

    erro_padrao = desvio_padrao / np.sqrt(n)
    t_critico = t.ppf(nivel_confianca, df=n-1) 

    limite_inferior = media - t_critico * erro_padrao
    limite_superior = media + t_critico * erro_padrao

    print(f"Média da amostra: {media:.3f}")
    print(f"Desvio padrão da amostra: {desvio_padrao:.3f}")
    print(f"Intervalo de confiança de: ({limite_inferior:.3f}, {limite_superior:.3f})")

    return (limite_inferior, limite_superior)
```

```python
# 95%
níveis_de_confiança_1 = calcular_intervalo_confianca(amostra, nivel_confianca=0.975)
```

```python
# 90%
níveis_de_confiança_2 = calcular_intervalo_confianca(amostra, nivel_confianca=0.95)
```

```python
# 99%
níveis_de_confiança_3 = calcular_intervalo_confianca(amostra, nivel_confianca=0.995)
```

```python
plt.figure(figsize=(8, 2))

y_val = 1  # posição fixa no eixo y para representar a qualidade visual

# Média como ponto
plt.scatter(media, y_val, color='red', label='Média')

plt.scatter(media, y_val*1.3, color='red')
plt.hlines(y=y_val*1.3, xmin=níveis_de_confiança_2[0], xmax=níveis_de_confiança_2[1], colors='green', linewidth=3, label='IC 90%')

# Linha do intervalo de confiança
plt.hlines(y=y_val, xmin=níveis_de_confiança_1[0], xmax=níveis_de_confiança_1[1], colors='gray', linewidth=3, label='IC 95%')

plt.scatter(media, y_val*0.7, color='red')
plt.hlines(y=y_val*0.7, xmin=níveis_de_confiança_3[0], xmax=níveis_de_confiança_3[1], colors='blue', linewidth=3, label='IC 99%')


# Ajustes do gráfico
plt.yticks([y_val], ['Qualidade Visual'])
plt.xlabel('Nota (0 a 1)')
plt.title('Média e Intervalo de Confiança de 95% para Qualidade Visual')
plt.xlim(0, 1)
plt.ylim(0.5, 1.5)
plt.legend(loc='upper center')
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# C: p-value em um Teste de Hipótese

```python
import numpy as np
import scipy.stats as stats

sample_size = 30
true_mean_1 = 20
true_std = 10
true_mean_2 = 70

# Gerando duas amostras 
np.random.seed(42)
group1 = np.random.normal(true_mean_1, true_std, sample_size)
group2 = np.random.normal(true_mean_2, true_std, sample_size)

# Aplicando o teste t
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"P-valor: {p_value:.2e}")
```

Hipótese nula (H₀): As médias dos dois grupos são iguais.

Hipótese alternativa (H₁): As médias dos dois grupos são diferentes.



nível de significancia = 0.05


R= O resultado do p_value mostra que há uma diferença significativa alta (3.05e-29 é muito menor que o nível de significância de 0.05), rejeitando a hipótese nula. Isso indica que os grupos têm diferenças. 

# D: significance_level

```python
import numpy as np
from scipy.stats import ttest_ind

def simulate_errotipoI(significance):
    num_simulations = 10000
    sample_size = 30
    true_mean = 50
    true_std = 10
    significance_level = significance#Usar valores diferentes
    
    
    significant_results = 0
    
    
    for _ in range(num_simulations):
    
        group1 = np.random.normal(true_mean, true_std, sample_size)
        group2 = np.random.normal(true_mean, true_std, sample_size)
    
    
        t_stat, p_value = ttest_ind(group1, group2)
    
    
        if p_value < significance_level:
            significant_results += 1
    
    
    proportion_significant = significant_results / num_simulations
    
    
    print(f"Number of simulations: {num_simulations}")
    print(f"Significance level (alpha): {significance_level}")
    print(f"Proportion of significant results (Type I error rate): {proportion_significant:.4f}")
```

```python
simulate_errotipoI(0.05)
```

```python
simulate_errotipoI(0.01)
```

## intervalo de confiança para população dados numéricos e binarios

Dados numéricos

```python
import numpy as np
import pandas as pd
from scipy.stats import t

# Semente para reprodutibilidade
np.random.seed(42)

# População A: dados numéricos (distribuição normal)
amostra_A = np.random.normal(loc=50, scale=10, size=30)  # média 50, desvio padrão 10


# População B: dados binários (distribuição binomial)
n= 1 # Total
p= 0.7 # Probabilidade de obter
amostra_B = np.random.binomial(n, p, size=30)  # 0 ou 1, com probabilidade de 0.7 para 1
```

```python
import matplotlib.pyplot as plt

def plot_ic(media, limite_inferior, limite_superior, y_val=1):
    # Criar gráfico
    plt.figure(figsize=(8, 2))

    # Média como ponto
    plt.scatter(media, y_val, color='red', label='Média')

    # Linha do intervalo de confiança
    plt.hlines(y=y_val, xmin=limite_inferior, xmax=limite_superior, colors='black', linewidth=3, label='IC 95%')

    # Ajustes do gráfico
    plt.yticks([y_val], [' '])
    plt.xlabel('')
    plt.title('Média e Intervalo de Confiança de 95%')
    plt.xlim(limite_inferior - (limite_inferior*0.1), limite_superior + (limite_inferior*0.1))
    plt.ylim(0.5, 1.5)
    plt.legend(loc='upper left')
    plt.grid(axis='x', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()
```

```python
# Calcular média e desvio padrão amostral
media = np.mean(amostra_A)
desvio_padrao = np.std(amostra_A, ddof=1)
n = len(amostra_A)

# Calcular intervalo de confiança de 95%  
t_critico = t.ppf(0.975, df=n-1)  


erro_padrao = desvio_padrao / np.sqrt(n)
limite_inferior = media - t_critico * erro_padrao
limite_superior = media + t_critico * erro_padrao

print(f"Média da amostra: {media:.3f}")
print(f"Desvio padrão da amostra: {desvio_padrao:.3f}")
print(f"t critico: {t_critico:.3f}")
print(f"erro padrao: {erro_padrao:.3f}")
print(f"Intervalo de confiança de: ({limite_inferior:.3f}, {limite_superior:.3f})")

plot_ic(media, limite_inferior, limite_superior)
```

```python
from scipy.stats import norm

mediaz = np.mean(amostra_B)
desvio_padraoz = np.std(amostra_B, ddof=1)
# Tamanho da amostra
n = len(amostra_B)

# Proporção de (1s) na amostra
p_s = np.mean(amostra_B)  
# Cálculo do erro padrão para proporções
# Fórmula: erro_padrao = sqrt(p̂ * (1 - p̂) / n)
produto = p_s * (1 - p_s)      
divisao = produto / n                
erro_padraoz = np.sqrt(divisao) 

# Valor crítico da normal padrão para 95% de confiança
z_critico = norm.ppf(0.975) 


limite_inferiorz = (p_s - z_critico * erro_padraoz)
limite_superiorz = (p_s + z_critico * erro_padraoz)

print(f"Média da amostra: {mediaz:.3f}")
print(f"Desvio padrão da amostra: {desvio_padraoz:.3f}")
print(f"z critico: {z_critico:.3f}")
print(f"erro padrao: {erro_padraoz:.3f}")
print(f"Intervalo de confiança de: ({limite_inferiorz:.3f}, {limite_superiorz:.3f})")

plot_ic(mediaz, limite_inferiorz, limite_superiorz)
```
