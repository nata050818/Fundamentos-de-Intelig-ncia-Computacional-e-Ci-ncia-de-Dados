# A - Correlação

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# pontos rosas y varia de 1 a 9 x fixo em 0
x_pink = np.zeros(9)  
y_pink = np.linspace(1, 9, 9)

# pontos verdes x varia de 1 a 9 y fixo em 0
x_green = np.linspace(1, 9, 9) 
y_green = np.zeros(9)  


# Regressão linear aplicada aos pontos azuis
slope, intercept, r_value, p_value, std_err = linregress(x_green, y_pink)
# slope * x_green + intercept, está gerando um array de valores y para cada ponto de x_green, que corresponde 
# a pontos ao longo da reta de regressão.
regression_line = slope * x_green + intercept  


fig, ax = plt.subplots(figsize=(8, 6))
# Plotando os pontos rosa
ax.scatter(x_pink, y_pink, color='pink', edgecolor='k', s=200, label='Gene Y variando (x=0)')
# Plotando os pontos verdes
ax.scatter(x_green, y_green, color='lightgreen', edgecolor='k', s=200, label='Gene X variando (y=0)')


# Plotando os pontos azuis: relação linear entre Gene X e Gene Y
ax.scatter(x_green, y_pink, color='skyblue', edgecolor='k', s=200, label='Cruzamento (x,y)')
# Plotando a reta de regressão baseada nos pontos azuis
ax.plot(x_green, regression_line, 'k--', label='Regressão linear')




ax.set_xlabel('Gene X', fontsize=14, weight='bold')
ax.set_ylabel('Gene Y', fontsize=14, weight='bold')
ax.legend()
plt.tight_layout()  # Ajusta layout para não cortar texto
plt.show()  # Exibe o gráfico

# Exibindo resultados da regressão
print(f"Inclinação: {slope:.2f}")
print(f"Intercepto: {intercept:.2f}")
print(f"Coeficiente de correlação (r): {r_value:.2f}")
print(f"P-valor: {p_value:.2e}")
print(f"Erro padrão: {std_err:.2f}")

```

A inclinação de 1.00 significa que, sempre que o valor de x aumenta 1 unidade, o valor de y também aumenta exatamente 1 unidade.

O intercepto de 0.00 mostra quando x é 0, y também é 0.

O coeficiente de correlação (r) igual a 1.00 indica uma relação perfeitamente linear e positiva.

O p-valor é extremamente pequeno (3.29 × 10⁻⁷⁰), o que confirma que essa relação não aconteceu por acaso (é altamente significativa).

O erro padrão é 0.00 o que significa que não há incerteza na inclinação da reta 

```python
# pontos rosas y varia de 1 a 9 x fixo em 0
x_pink = np.zeros(9)  
y_pink = np.random.uniform(1, 9, 9)

# pontos verdes x varia de 1 a 9 y fixo em 0
x_green = np.random.uniform(1, 9, 9) 
y_green = np.zeros(9)  


# Regressão linear aplicada aos pontos azuis
slope, intercept, r_value, p_value, std_err = linregress(x_green, y_pink)
# slope * x_green + intercept, está gerando um array de valores y para cada ponto de x_green, que corresponde 
# a pontos ao longo da reta de regressão.
regression_line = slope * x_green + intercept  


fig, ax = plt.subplots(figsize=(8, 6))
# Plotando os pontos rosa
ax.scatter(x_pink, y_pink, color='pink', edgecolor='k', s=200, label='Gene Y variando (x=0)')
# Plotando os pontos verdes
ax.scatter(x_green, y_green, color='lightgreen', edgecolor='k', s=200, label='Gene X variando (y=0)')


# Plotando os pontos azuis: relação linear entre Gene X e Gene Y
ax.scatter(x_green, y_pink, color='skyblue', edgecolor='k', s=200, label='Cruzamento (x,y)')
# Plotando a reta de regressão baseada nos pontos azuis
ax.plot(x_green, regression_line, 'k--', label='Regressão linear')




ax.set_xlabel('Gene X', fontsize=14, weight='bold')
ax.set_ylabel('Gene Y', fontsize=14, weight='bold')
ax.legend()
plt.tight_layout()  # Ajusta layout para não cortar texto
plt.show()  # Exibe o gráfico

# Exibindo resultados da regressão
print(f"Inclinação: {slope:.2f}")
print(f"Intercepto: {intercept:.2f}")
print(f"Coeficiente de correlação (r): {r_value:.2f}")
print(f"P-valor: {p_value:.2e}")
print(f"Erro padrão: {std_err:.2f}")

```

A inclinação da reta de regressão é -0.25, o que indica uma tendência negativa. A medida que o valor de x aumenta, o valor de y tende a diminuir (inclinação é pequena, tendência é fraca).

O intercepto é 5.12, o que representa o valor estimado de y quando x é igual a zero. Esse valor serve como ponto de partida da reta.

O coeficiente de correlação (r) é -0.22, indicando uma correlação negativa fraca entre x e y. 

O p-valor foi (5.61 × 10⁻¹), o que é muito acima do nível de significância (0.05). Isso mostra que a inclinação observada pode facilmente ter ocorrido por acaso (Não é estatisticamente significativa).

O erro padrão da inclinação é 0.41, maior do que o valor da própria inclinação (-0.25), indicando grande incerteza na estimativa.

linregress(): Calcula a inclinação (slope) e o intercepto da reta que minimiza os erros quadráticos (método dos mínimos quadrados).
